# IMPORT DEPENDENCIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.model_selection import train_test_split
```

# DATASET

```python
data = pd.read_csv('housing.csv')
print(data)
```

```python
print(data.info())
```

```python
data.dropna(inplace = True)
print(data)
```

```python
print(data.info())
```

```python
x = data.drop(['median_house_value'], axis = 1)
y = data['median_house_value']
```

```python
print(x,y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
tr_data = X_train.join(y_train)
print(tr_data)
tr_data.hist(figsize = (15,10))
tr_data.plot(figsize = (15,10))
```

```python
tr_data.corr()
```

```python
plt.figsize = (15,10)
sns.heatmap(tr_data.corr(), annot= True, cmap = "RdYlBu")
```

```python
tr_data['total_rooms'] = np.log(tr_data['total_rooms'] + 1)
tr_data['total_bedrooms'] = np.log(tr_data['total_bedrooms'] + 1)
tr_data['population'] = np.log(tr_data['population'] + 1)
tr_data['households'] = np.log(tr_data['households'] + 1)
```

```python
tr_data.hist(figsize = (15,10))
```

```python
tr_data.ocean_proximity.value_counts()
```

```python
tr_data = tr_data.join(pd.get_dummies(tr_data.ocean_proximity)).drop(['ocean_proximity'], axis = 1)
```

```python
tr_data
```

```python
plt.figure(figsize = (15,10))
sns.scatterplot(x = 'latitude',y = 'longitude',data = tr_data,hue = 'median_house_value',palette = 'coolwarm')
## blue == less expensive ,, red == more expensive
```

```python
tr_data['bed_per_room'] = tr_data['total_bedrooms'] / tr_data['total_rooms']
tr_data['room_per_household'] = tr_data['total_rooms'] / tr_data['households']
```

```python
plt.figsize = (15,8)
sns.heatmap(tr_data.corr(), annot= True, cmap = "RdYlBu")
```

```python
from sklearn.linear_model import LinearRegression
x_train,y_train = tr_data.drop(['median_house_value'],axis = 1), tr_data['median_house_value']
lr = LinearRegression()
lr.fit(x_train,y_train)
```

#  Test data

```python
test_data = X_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'], axis = 1)

test_data['bed_per_room'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['room_per_household'] = test_data['total_rooms'] / test_data['households']
```

```python
x_test,y_test = test_data.drop(['median_house_value'],axis = 1), test_data['median_house_value']
```

```python
lr.score(x_test,y_test)
```
